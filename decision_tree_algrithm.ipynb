```python
import pandas as pd
df=pd.read_csv("/content/titanic.csv")
df.head()
```

```python
inp=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
inp.head()
```

```python
target=df['Survived']
target
```

```python
from sklearn.preprocessing import LabelEncoder
el_pc=LabelEncoder()
el_sex=LabelEncoder()
el_age=LabelEncoder()
el_fare=LabelEncoder()
```

```python
inp['pc_n']=el_pc.fit_transform(inp['Pclass'])
inp['pc_sex']=el_pc.fit_transform(inp['Sex'])
inp['pc_age']=el_pc.fit_transform(inp['Age'])
inp['pc_fare']=el_pc.fit_transform(inp['Fare'])
inp
```

```python
inpn=inp.drop(['Pclass','Sex','Age','Fare','Survived'],axis='columns')
inpn
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
model.fit(inpn,target)
model.score(inpn,target)
```

```python
model.predict([[1,1,35,85]])
```
